```python
import pandas as pd
data = pd.read_csv('../data_clean/Afspraak_account_gelinkt_cleaned_fixed.csv', sep=",")
data.head()
```

```python
data[data['afspraak_account_gelinkt_keyphrases'].str.len() > 2000]['afspraak_account_gelinkt_keyphrases']
```

```python
data["afspraak_account_gelinkt_keyphrases"] = data["afspraak_account_gelinkt_keyphrases"].str.split(",")

  # Split each sentence by space to get an array of words within each sentence
data["afspraak_account_gelinkt_keyphrases"] = data["afspraak_account_gelinkt_keyphrases"].apply(
    lambda sentence_list: [sentence.split() if isinstance(sentence, str) else sentence for sentence in sentence_list]
)

data.head()
```

```python
import tensorflow_hub as hub
from tqdm import tqdm
import pandas as pd

# Load the Universal Sentence Encoder
embed = hub.load("https://tfhub.dev/google/universal-sentence-encoder/4")

row = []

for sentences in tqdm(data["afspraak_account_gelinkt_keyphrases"]):
    sentence_embedding = []
    for sentence in sentences:
        word_embedding = []
        for word in sentence:
            embedding = embed([word])
            word_embedding.append(embedding[0].numpy()) 
        # Convert the list of word embeddings to a string and remove "array(" and "dtype=float32)"
    row.append(sentence_embedding)

# Extract and convert the embeddings to NumPy arrays
data["afspraak_account_gelinkt_keyphrases_embedding"] = row
```

```python
import pandas as pd
import re

data = pd.read_csv('../data/cdi pageviews.csv', encoding="latin-1", sep=";")
data.columns = data.columns.map(lambda x: re.sub(r'^crm CDI_PageView\[(.*)\]$', r'\1', x))
data.columns = data.columns.map(lambda x: re.sub(r' ', r'_', x))
data.drop(['Anonymous_Visitor'], axis=1, inplace=True)
data.head()
```

```python
import os
import pandas as pd

directory_path = "../data/"
for file in os.listdir(directory_path):
    if file.endswith(".csv"):
        try:
            file_path = os.path.join(directory_path, file)
            df = pd.read_csv(file_path, encoding="latin-1", sep=";")
            num_rows = len(df)
            print(f"File: {file}, Number of rows: {num_rows}")
            print(df.columns[df.isnull().all()])
            print(*50*"-", sep="")
        except:
            print(f"File: {file} could not be read")
            print(*50*"-", sep="")
```

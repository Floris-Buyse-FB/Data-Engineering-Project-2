## Data Cleaning

```python
# combine all the data 
df3= df.copy()
df2 =df
df2['data'] =df[df.columns[1:]].apply(lambda x: ','.join(x.dropna().astype(str)),axis=1)
print(df['data'].head())
```

```python
from sklearn.feature_extraction.text import CountVectorizer

vectorizer = CountVectorizer()
vectorized= vectorizer.fit_transform(df2['data'])
```

```python
from sklearn.metrics.pairwise import cosine_similarity

#duurt 2m38s
similarities = cosine_similarity(vectorized)
```

```python
print(similarities)
```

```python
df = pd.DataFrame(similarities,columns=df['contact_contactpersoon_id'],index=df['contact_contactpersoon_id']).reset_index()
df.head()
```

```python
input_person_id = 'D9303EA2-57E5-EB11-8121-001DD8B72B61'
recommendations = pd.DataFrame(df.nlargest(11,input_person_id)['contact_contactpersoon_id'])
recommendations = recommendations[recommendations['contact_contactpersoon_id']!=input_person_id]
print(recommendations)
```

```python
import pandas as pd
import numpy as np
import os
import warnings
warnings.filterwarnings('ignore')
```

```python
DATA_DIR = os.path.join(os.path.abspath(os.path.join(os.getcwd(), '..')), 'data_clean')
DATA_DIR
```

```python
for file in os.listdir(DATA_DIR):
    if file.endswith(".csv") and (not file.endswith("_Merge.csv")):

        print(f"--{file}")
        df = pd.read_csv(os.path.join(DATA_DIR, file), sep=',')
        df.columns = df.columns.str.replace(' ', '_')

        sql_data_types = {
            'int64': 'INT',
            'float64': 'FLOAT',
            'object': 'VARCHAR(255)',
            'datetime64[ns]': 'DATE',
        }
        table_name = file[:-10]  # Remove the '.csv' extension to use as the table name
        table_name = table_name.replace(' ', '_')

        columns = ',\n'.join([f'{col} {sql_data_types.get(str(dtype), "VARCHAR(255)")}' for col, dtype in zip(df.columns, df.dtypes)])
        create_table_query = f'IF NOT EXISTS (SELECT * FROM INFORMATION_SCHEMA.TABLES WHERE TABLE_NAME = \'{table_name}\')\nBEGIN\nCREATE TABLE {table_name} (\n{columns});\nEND'
        
        # Print the generated CREATE TABLE query
        print(create_table_query)
        print()
```
